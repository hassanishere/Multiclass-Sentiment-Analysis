```python
import pandas as pd

# Load the CSV files
roberta_df = pd.read_csv('/content/drive/My Drive/lastresults/RoBERTa_validation_with_predictions_final .csv')
bert_df = pd.read_csv('/content/drive/My Drive/lastresults/BERT_validation_with_predictions_final.csv')
gpt2_df = pd.read_csv('/content/drive/My Drive/lastresults/GPT2_Final_Validation (1).csv')
distilbert_df = pd.read_csv('/content/drive/My Drive/lastresults/DistilBERT_validation_with_predictions_final.csv')
```

```python
import pandas as pd


distilbert_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
distilbert_df


roberta_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
roberta_df

bert_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
bert_df

gpt2_df.rename(columns={'Review Text': 'reviewText'}, inplace=True)

```

```python
# Display the first few observations for each DataFrame
print("RoBERTa Data:")
print(roberta_df.head())

print("\nBERT Data:")
print(bert_df.head())

print("\nGPT-2 Data:")
print(gpt2_df.head())

print("\nDistilBERT- Data:")
print(distilbert_df.head())


```

```python
import pandas as pd

# Function to map numerical labels to polarity strings
def map_polarity(label):
    if label in [0, 1]:
        return 'negative'
    elif label == 2:
        return 'neutral'
    elif label in [3, 4]:
        return 'positive'

# Function to add actual_polarity and predicted_polarity columns
def add_polarity_columns(df):
    df['actual_polarity'] = df['Actual Label'].apply(map_polarity)
    df['predicted_polarity'] = df['Predicted Label'].apply(map_polarity)
    return df

# Apply the function to each DataFrame
bert_df = add_polarity_columns(bert_df)
roberta_df = add_polarity_columns(roberta_df)
distilbert_df = add_polarity_columns(distilbert_df)
gpt2_df = add_polarity_columns(gpt2_df)

# Show the updated DataFrames
print("Updated bert_df:")
print(bert_df)

print("\nUpdated roberta_df:")
print(roberta_df)

print("\nUpdated distilbert_df:")
print(distilbert_df)

print("\nUpdated gpt2_df:")
print(gpt2_df)
```

```python
# Create new DataFrames for misclassified rows
misclassified_bert = bert_df[bert_df['actual_polarity'] != bert_df['predicted_polarity']]
misclassified_roberta = roberta_df[roberta_df['actual_polarity'] != roberta_df['predicted_polarity']]
misclassified_distilbert = distilbert_df[distilbert_df['actual_polarity'] != distilbert_df['predicted_polarity']]
misclassified_gpt2 = gpt2_df[gpt2_df['actual_polarity'] != gpt2_df['predicted_polarity']]

# Show the number of rows in each misclassified DataFrame
print("Number of rows in misclassified_bert:", len(misclassified_bert))
print("Number of rows in misclassified_roberta:", len(misclassified_roberta))
print("Number of rows in misclassified_distilbert:", len(misclassified_distilbert))
print("Number of rows in misclassified_gpt2:", len(misclassified_gpt2))

# Show the misclassified DataFrames
print("\nMisclassified bert_df:")
print(misclassified_bert)

print("\nMisclassified roberta_df:")
print(misclassified_roberta)

print("\nMisclassified distilbert_df:")
print(misclassified_distilbert)

print("\nMisclassified gpt2_df:")
print(misclassified_gpt2)
```

```python
# Merge the misclassified DataFrames on reviewText, actual_polarity, and predicted_polarity
common_misclassifications = misclassified_bert.merge(misclassified_roberta, on=['reviewText', 'actual_polarity', 'predicted_polarity'], how='inner', suffixes=('_bert', '_roberta'))
common_misclassifications = common_misclassifications.merge(misclassified_distilbert, on=['reviewText', 'actual_polarity', 'predicted_polarity'], how='inner', suffixes=('_common', '_distilbert'))
common_misclassifications = common_misclassifications.merge(misclassified_gpt2, on=['reviewText', 'actual_polarity', 'predicted_polarity'], how='inner', suffixes=('_common', '_gpt2'))

# Show the number of rows in the common misclassifications DataFrame
print("Number of rows in common misclassifications:", len(common_misclassifications))

# Show the common misclassifications DataFrame
print("\nCommon misclassifications:")
print(common_misclassifications)
```

```python
# Selecting only the desired columns
final_common_misclassifications = common_misclassifications[['reviewText', 'actual_polarity', 'predicted_polarity']]

# Show the final_common_misclassifications DataFrame
print("final_common_misclassifications:")
print(final_common_misclassifications)
```

```python
# Merge misclassified_gpt2 and final_common_misclassifications based on reviewText with a left join
merged_diff_gpt2 = misclassified_gpt2.merge(final_common_misclassifications, on='reviewText', how='left', indicator=True)

# Filter the merged DataFrame to show only the rows where final_common_misclassifications has a null value
rows_not_in_common = merged_diff_gpt2[merged_diff_gpt2['_merge'] == 'left_only']

# Selecting only the desired columns from rows_not_in_common
gpt2_rows_not_in_common = rows_not_in_common[['reviewText', 'Actual Label', 'Predicted Label']]

# Show the rows in misclassified_roberta that are not present in final_common_misclassifications
print("Rows in misclassified_roberta that are not present in final_common_misclassifications based on reviewText:")
print(gpt2_rows_not_in_common)
```
