```python
import pandas as pd


distilbert_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
distilbert_df


roberta_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
roberta_df

bert_df.rename(columns={'overall': 'Actual Label', 'predicted_values': 'Predicted Label'}, inplace=True)
bert_df

gpt2_df.rename(columns={'Review Text': 'reviewText'}, inplace=True)

```

```python
gpt2_df
```

```python
# List of DataFrames with a corresponding label
dfs = [('bert_df', bert_df), ('roberta_df', roberta_df), ('distilbert_df', distilbert_df), ('gpt2_df', gpt2_df)]

# Filter rows in each DataFrame where reviewText starts with the specified text and add a new column indicating the source DataFrame
filtered_dfs = [(df_name, df[df['reviewText'].str.startswith('The title of this review')]) for df_name, df in dfs]

# Concatenate the filtered DataFrames
result_df = pd.concat([df.assign(source_df=df_name) for df_name, df in filtered_dfs])

# Display the result
print(result_df)
```
